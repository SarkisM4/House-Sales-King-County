# House Sales in King County, USA


This dataset contains house sale prices from King County, in Seattle. It includes homes sold between May 2014 and May 2015.


| Variable      | Description                                                                                                 |
| ------------- | ----------------------------------------------------------------------------------------------------------- |
| id            | A notation for a house                                                                                      |
| date          | Date house was sold                                                                                         |
| price         | Price is prediction target                                                                                  |
| bedrooms      | Number of bedrooms                                                                                          |
| bathrooms     | Number of bathrooms                                                                                         |
| sqft_living   | Square footage of the home                                                                                  |
| sqft_lot      | Square footage of the lot                                                                                   |
| floors        | Total floors (levels) in house                                                                              |
| waterfront    | House which has a view to a waterfront                                                                      |
| view          | Has been viewed                                                                                             |
| condition     | How good the condition is overall                                                                           |
| grade         | overall grade given to the housing unit, based on King County grading system                                |
| sqft_above    | Square footage of house apart from basement                                                                 |
| sqft_basement | Square footage of the basement                                                                              |
| yr_built      | Built Year                                                                                                  |
| yr_renovated  | Year when house was renovated                                                                               |
| zipcode       | Zip code                                                                                                    |
| lat           | Latitude coordinate                                                                                         |
| long          | Longitude coordinate                                                                                        |
| sqft_living15 | Living room area in 2015(implies-- some renovations) This might or might not have affected the lotsize area |
| sqft_lot15    | LotSize area in 2015(implies-- some renovations)                                                            |

## Cleaning the Data


#### Dropping the columns <code>"id"</code>  and <code>"Unnamed: 0"</code> from axis 1 using the method <code>drop()</code>, then using the method <code>describe()</code> to obtain a statistical summary of the data. 


```python
df.drop(['id', 'Unnamed: 0'], axis = 1, inplace = True)
df.describe()
```

We can see we have missing values for the columns <code> bedrooms</code>  and <code> bathrooms </code>


```python
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

We can replace the missing values of the column <code>'bedrooms'</code> with the mean of the column  <code>'bedrooms' </code> using the method <code>replace()</code>.


```python
mean = df['bedrooms'].mean()
df['bedrooms'].replace(np.nan, mean, inplace = True)
```

We also replace the missing values of the column <code>'bathrooms'</code> with the mean of the column  <code>'bathrooms' </code> using the method <code>replace()</code>.


```python
mean=df['bathrooms'].mean()
df['bathrooms'].replace(np.nan,mean, inplace=True)
```

```python
print("number of NaN values for the column bedrooms :", df['bedrooms'].isnull().sum())
print("number of NaN values for the column bathrooms :", df['bathrooms'].isnull().sum())
```

#### Using the  <code>value_counts</code> method to count the number of houses with unique floor values, use the method <code>.to_frame()</code> to convert it to a dataframe.


```python
df["floors"].value_counts().to_frame()
```

## Visualizing the Data

#### Using the function <code>boxplot</code> from the seaborn library to determine whether houses with a waterfront view or without a waterfront view have more price outliers.


```python
sns.boxplot(x = 'waterfront', y = 'price', data = df)
```

The Boxplot for Houses with a waterfront view are labeled with 1 on the x-axis, and the boxplot for houses without a waterfront view are labeled with a 0 on the x-axis. Based on the boxplots we can see that houses without a waterfront view appear to have more price outliers than the houses with a waterfront view.

#### Using the function <code>regplot</code>  from the seaborn library to determine if the feature <code>sqft_above</code> is negatively or positively correlated with <code>price</code>.


```python
sns.regplot(x = 'sqft_above', y = 'price', data = df)
plt.ylim(0,)
```

Based on the regression plot, there does appear to be a somewhat strong and postive correlation between the sqft of the house and the price.

We can use the Pandas method <code>corr()</code>  to find the feature,other than price, that is most correlated with price.
